```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.figure_factory as ff

#Implementation of hierarchical clustering(Dendogram)
def dendogram_plot(means, data_sizes,cov, title):
    data = []
    classes = len(means)
    for i in range(classes):
        x, y = np.random.multivariate_normal(means[i], cov, data_sizes[i]).T
        for j in range(len(x)):
            data.append([x[j],y[j]])
    fig = ff.create_dendrogram(np.array(data))
    fig.update_layout(width=800, height=500, title = title)
    fig.show()

    #i=0
    #for point in data:
    #    print("point - "+str(i)+ " : "+str(point[0])+","+str(point[1]))
    #    i+=1


neg_cov = [[1, -0.7], [-0.7, 1]]
pos_cov = [[1, 0.7], [0.7, 1]]
```

```python
#Input means for distinct 
means = [[0,0],[5,5]]
#Input data sizes for distinct
data_sizes = [10,10]
dendogram_plot(means, data_sizes, pos_cov, "Distinct & Positive Covariance Dendogram")
```

```python
#Input means for partial overlap
means = [[0,0],[3,3]]
#Input data sizes for partial overlap
data_sizes = [10,10]
dendogram_plot(means, data_sizes, pos_cov, "Partial Overlap Dendogram - Positive Covariance")
```

```python
#Input means for fully overlap
means = [[0,0],[0.5,0.5]]
#Input data size for fully overlap
data_sizes = [10,10]
dendogram_plot(means, data_sizes, pos_cov, "Fully Overlap Dendogram - Positive Co-variance")
```

```python
#Balanced data with multiple classes
means = [[0,0],[1,1],[5,5],[5,5]]
data_sizes = [10,10,10,10]
dendogram_plot(means, data_sizes, neg_cov, "4 classes with balanced data")
```
