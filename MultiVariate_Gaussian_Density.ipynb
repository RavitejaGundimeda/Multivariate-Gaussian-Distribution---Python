A program to generate Random Dot distributions using Multivariate Gaussian 
(Normal) Density. In two dimensions these will appear to be elliptical in shape, or in 3D 
as ellipsoids. The position of Mean Vector and **the** Covariance allows you to control the 
distributions. Use different coloured dots to show different classes. Make plots for 
different degree of overlap between two classes, like totally distinct and separate, partial 
overlap, full overlap. You should be able to iterate your plots for C different class, C is a 
non-zero integer. The number of data in each class can vary experiment with equal 
numbers and different types of imbalances.

(i) Implementing Multivariate Gaussian(Normal) Density:


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.figure_factory as ff

def multi_class_norm_plot(means, data_sizes, cov, title):
    colours = ['b','c','r','g','m']
    classes = len(means)
    for i in range(classes):
        x, y = np.random.multivariate_normal(means[i], cov, data_sizes[i]).T
        plt.title(title)
        plt.plot(x, y,'x',colours[i], alpha = 0.5)

neg_cov = [[1, -0.7], [-0.7, 1]]
pos_cov = [[1, 0.7], [0.7, 1]]
```

(ii) Implementing different degree of overlap between two classes with positive 
covariance.

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(5)

means = [[0,0],[5,5]]
data_sizes = [500,500]
plt.subplot(1,3,1)
multi_class_norm_plot(means, data_sizes, pos_cov, "Positive Covariance-No Overlap")

means = [[0,0],[3,3]]
data_sizes = [500,500]
plt.subplot(1,3,2)
multi_class_norm_plot(means, data_sizes, pos_cov, "Positive Covariance-Partial Overlap")

means = [[0,0],[0.5,0.5]]
data_sizes = [500,500]
plt.subplot(1,3,3)
multi_class_norm_plot(means, data_sizes, pos_cov, "Positive Covariance-Fully Overlap")
plt.show()
```

(iii) Implementing different degree of overlap between two classes with negative 
covariance

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(5)

means = [[0,0],[5,5]]
data_sizes = [500,500]
plt.subplot(1,3,1)
multi_class_norm_plot(means, data_sizes, neg_cov, "Negative Covariance & No Overlap")

means = [[0,0],[1.25,1.25]]
data_sizes = [500,500]
plt.subplot(1,3,2)
multi_class_norm_plot(means, data_sizes, neg_cov, "Negative Covariance & Partial Overlap")

means = [[0,0],[0.5,0.5]]
data_sizes = [500,500]
plt.subplot(1,3,3)
multi_class_norm_plot(means, data_sizes, neg_cov, "Negative Covariance & Fully Overlap")
plt.show()
```

(iv) Implementing Multivariate for C(=4) different classes with balanced and 
Imbalanced data size

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(5)

means = [[0,0],[1.5,1.5],[5,5],[5.5,5.5]]
data_sizes = [500,500,500,500]
plt.subplot(1,3,1)
multi_class_norm_plot(means, data_sizes, neg_cov, "4 Different Classes with Balanced size")

means = [[0,0],[1.5,1.5],[5,5],[5.5,5.5]]
data_sizes = [500,100,1000,200]
plt.subplot(1,3,2)
multi_class_norm_plot(means, data_sizes, neg_cov, "4 Different Classes with Imbalanced Size")
```
